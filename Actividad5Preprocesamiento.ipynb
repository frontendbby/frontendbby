3. Clasifique la distribución del peso según su asimetría, justificando tu respuesta. Incluye el cálculo numérico que lo demuestra.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, norm

response_df_clean = response_df['Weight'].dropna()

weight_skewness = skew(response_df_clean)
print(f'Coeficiente de asimetría: {weight_skewness}')

if weight_skewness > 0:
    print("La distribución del peso está sesgada a la derecha.")
elif weight_skewness < 0:
    print("La distribución del peso está sesgada a la izquierda.")
else:
    print("La distribución del peso es simétrica.")

plt.hist(response_df_clean, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

mean = np.mean(response_df_clean)
std_dev = np.std(response_df_clean)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)

plt.plot(x, p, 'k', linewidth=2)
plt.title('Distribución de Peso con Ajuste Normal')
plt.xlabel('Peso')
plt.ylabel('Densidad')
plt.grid()

plt.show()
```

```python
from scipy.stats import chi2_contingency
def Diagnose_MV_Categorical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    plt.figure(figsize=(10,4))
    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].value_counts().plot.bar()
        plt.title(MV_labels[poss])
    plt.show()

    contigency_table = pd.crosstab(BM_MV,df[str_att_name])
    p_value = chi2_contingency(contigency_table)[1]

    print('p-value of Chi_squared test: {}'.format(p_value))
```

6. Utiliza una medida de tendencia central para imputar los valores faltantes de `Weight`. ¿Cuál es recomendable y por qué?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

missing_percentage = response_df['Weight'].isnull().mean() * 100
print(f'Porcentaje de valores faltantes en "Weight": {missing_percentage:.2f}%')

missing_values = response_df.isnull().sum()
print("Valores faltantes en cada columna:")
print(missing_values[missing_values > 0])

response_df['Weight'].fillna(response_df['Weight'].mean(), inplace=True)

plt.hist(response_df['Weight'], bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

mean = response_df['Weight'].mean()
std_dev = response_df['Weight'].std()

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)

plt.plot(x, p, 'k', linewidth=2)

plt.title('Histograma de Peso con Ajuste Normal')
plt.xlabel('Peso')
plt.ylabel('Densidad')
plt.grid()

plt.show()
```
